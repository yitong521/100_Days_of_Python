```python
import pandas as pd
import re
```

```python
word_number_dict = {
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9
}
```

```python
file_path = 'F1P2puzzle.txt'
delimiter = ' '
df = pd.read_csv(file_path,header=None,  delimiter=delimiter)
df.columns = [ 'string']
```

```python
def find_first_and_last_digit(input_string):
    # 查找第一个数字
    first_digit_match = re.search(r'\d', input_string)
    first_digit = int(first_digit_match.group()) if first_digit_match else None

    # 查找最后一个数字
    last_digit_match = re.search(r'\d', input_string[::-1])
    last_digit = int(last_digit_match.group()) if last_digit_match else None

    return first_digit, last_digit

def combine_to_two_digit_number(num1, num2):
    # 将两个整数合并成字符串
    combined_str = str(num1) + str(num2)

    # 截取字符串的前两位（如果字符串长度超过两位）
    result_str = combined_str[:2]

    # 将结果转换为整数
    result_number = int(result_str)

    return result_number
def find_number_and_index(s, word_number_dict):
    i = 0
    result = ""
    index = -1  # Default value if no match is found
    
    while i < len(s):
        found_match = False
        
        for word, number in word_number_dict.items():
            if s[i:].startswith(word):
                result += str(number)
                index = i
                found_match = True
                break
        
        if found_match:
            break  # Exit the loop if a match is found
        else:
            i += 1
    
    return result, index
def find_last_number_reverse(s, word_number_dict):
    s_reverse = s[::-1]  # 反转字符串
    i = 0
    result = ""
    index = -1

    while i < len(s_reverse):
        found_match = False

        for word, number in word_number_dict.items():
            if s_reverse[i:].startswith(word[::-1]):
                index = len(s) - i - len(word)  # 计算匹配单词在原字符串中的索引
                result = str(number)
                found_match = True
                break

        if found_match:
            break  # 退出循环，因为只需要找到一个匹配项
        else:
            i += 1

    return result, index
def find_first_and_last_digit_with_index(input_string):
    # 查找第一个数字
    first_digit_match = re.search(r'\d', input_string)
    first_digit = int(first_digit_match.group()) if first_digit_match else None
    first_digit_index = first_digit_match.start() if first_digit_match else -1

    # 查找最后一个数字
    last_digit_match = re.search(r'\d', input_string[::-1])
    last_digit = int(last_digit_match.group()) if last_digit_match else None
    last_digit_index = len(input_string) - last_digit_match.start() - 1 if last_digit_match else -1

    return first_digit, first_digit_index, last_digit, last_digit_index
def compare_two_index_later(first_digit, first_digit_index,first_letternum, first_letternum_index):
    if first_digit_index == -1:
            first_digit_final = first_letternum
    elif first_letternum_index == -1:
        first_digit_final = first_digit
    elif first_digit_index > first_letternum_index:
        first_digit_final = first_digit
    else:
        first_digit_final = first_letternum
    return first_digit_final

def compare_two_index(first_digit, first_digit_index,first_letternum, first_letternum_index):
    if first_digit_index == -1:
            first_digit_final = first_letternum
    elif first_letternum_index == -1:
        first_digit_final = first_digit
    elif first_digit_index < first_letternum_index:
        first_digit_final = first_digit
    else:
        first_digit_final = first_letternum
    return first_digit_final
```
